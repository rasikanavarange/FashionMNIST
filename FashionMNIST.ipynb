```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

#### Import Dataset

```python
# load data
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

There are 60,000 images in the training set and 10,000 images in the test set. Each image is a 28x28 array of pixels

```python
# clothing class names
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

#### Preprocess Data

```python
for i in range(3):
    plt.figure()
    plt.imshow(train_images[i])
    plt.colorbar()
```

```python
# normalise the data
train_images = train_images / 255.0
test_images = test_images /255.0
```

```python
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
```

#### Building the Model

```python
model = keras.Sequential([
    # Transforms input images from 28x28 array to 1x784 array
    keras.layers.Flatten(input_shape=(28, 28)),
    # Fully connected layer with 128 neurons
    keras.layers.Dense(128, activation=tf.nn.relu),
    # Softmax layer, node for each class
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
# Compile the model
model.compile(optimizer=tf.train.AdamOptimizer(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```
